```python
import nltk
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords 


```

```python
pip install nltk
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Shahid_Afridi').read()
```

```python
soup = bs.BeautifulSoup(source, 'lxml')
```

```python
text = ""
for paragraph in soup.find_all('p'):
  text += paragraph.text
```

```python
text = re.sub(r'\[[0-9]*\]',' ',text)
```

```python
text
```

```python
text = re.sub(r'\n',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
```

```python
text
```

```python
text = re.sub(r'\s+',' ',text)
```

```python
text
```

```python
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences
```

```python
sentences1 = [nltk.word_tokenize(sentence) for sentence in sentences]
```
